```python
import os
os.chdir("../")
from dataset import Dataset
import matplotlib.pyplot as plt
import cv2
import numpy as np
%load_ext autoreload
%autoreload 2 

%matplotlib inline
```

```python
path = 'Training/'
img_w = 300
img_h = 300
n_labels = 2 
dataset = Dataset(path, img_w, img_h, n_labels)
image_name = '4.png'
annotation_name = '4-annotation.png'
x, y = dataset.single_image_augmentation(image_name, annotation_name)

img_titles=['origin', 'zoom and crop', 'zoom and crop', 'horizontal flip', 'rotation', 'zoom', 'motion blur', 'add', 'hue saturation', 'shear', 'inverted', 'sigmoid contrast']
imgs=[]

for img, title in zip(x, img_titles):
    violet= np.zeros((40, img.shape[1], 3), np.uint8)
    violet[:] = (255, 255, 255)
    violet=violet/255
    img = cv2.vconcat((violet, img))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(img,title,(10,30), font, 1,(0,0,0), 2, 0)
    imgs.append((img*255.0))
    

data_augmentation_image = dataset.display_data_augmentation(imgs, 4, 3, 10, 40)

plt.imshow(data_augmentation_image)
plt.show()

plt.imsave('./paper/paper_images/data_augmentation.png', data_augmentation_image)
```
